# Over-Sampling

```python
from imblearn.over_sampling import SMOTE
```

```python
smote=SMOTE('auto')
```

```python
x_sm,y_sm=smote.fit_sample(X_train,y_train)
```

```python
print(x_sm.shape,y_sm.shape)
```

```python
from sklearn.linear_model import LogisticRegression
classifier= LogisticRegression(random_state=0)
classifier.fit(x_sm,y_sm)

```

```python
y_pred=classifier.predict(X_test)

y_pred_train=classifier.predict(x_sm)

```

```python
from sklearn.metrics import confusion_matrix
cm= confusion_matrix(y_test, y_pred)

cm2= confusion_matrix(y_sm,y_pred_train)

print ("Confusion Matrix : \n", cm)

print ("Confusion Matrix : \n", cm2)
```

```python

from sklearn.metrics import accuracy_score 
print ("Accuracy : ", accuracy_score(y_test, y_pred)) 

from sklearn.metrics import accuracy_score 
print ("Accuracy : ", accuracy_score(y_sm, y_pred_train)) 

```

```python
from sklearn.metrics import roc_auc_score
print(roc_auc_score(y_test,y_pred))

print(roc_auc_score(y_sm,y_pred_train))

from sklearn.metrics import plot_roc_curve
import matplotlib.pyplot as plt

sklearn.metrics.plot_roc_curve(classifier, X_test, y_test)
```

# undersampling

```python
from imblearn.under_sampling import NearMiss 
nr = NearMiss()
```

```python
X_train_miss, y_train_miss = nr.fit_sample(X_train, y_train)
```

```python
print(X_train_miss.shape,X_train_miss.shape)
```

```python
from sklearn.linear_model import LogisticRegression
classifier= LogisticRegression(random_state=0)
classifier.fit(X_train_miss, y_train_miss)

```

```python
y_pred=classifier.predict(X_test)

y_pred_train=classifier.predict(X_train_miss)

```

```python
from sklearn.metrics import confusion_matrix
cm= confusion_matrix(y_test, y_pred)

cm2= confusion_matrix(y_train_miss,y_pred_train)

print ("Confusion Matrix : \n", cm)

print ("Confusion Matrix : \n", cm2)

```

```python

 
print ("Accuracy : ", accuracy_score(y_test, y_pred)) 

 
print ("Accuracy : ", accuracy_score(y_train_miss, y_pred_train)) 
```
